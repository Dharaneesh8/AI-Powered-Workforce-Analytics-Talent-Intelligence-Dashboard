Create a e-commerce transaction data with 10 rows and 10 columns

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the number of rows and columns
num_rows = 10
num_columns = 10

# Generate random data for e-commerce transactions
data = {
    'TransactionID': np.random.randint(10000, 99999, num_rows),
    'CustomerID': np.random.randint(1000, 9999, num_rows),
    'ProductID': np.random.randint(100, 999, num_rows),
    'Quantity': np.random.randint(1, 10, num_rows),
    'Price': np.round(np.random.uniform(5.0, 500.0, num_rows), 2),
    'TransactionDate': [datetime.now() - timedelta(days=np.random.randint(0, 30)) for _ in range(num_rows)],
    'PaymentMethod': np.random.choice(['Credit Card', 'PayPal', 'Bank Transfer'], num_rows),
    'ShippingAddress': [f'Address {i+1}' for i in range(num_rows)],
    'Status': np.random.choice(['Completed', 'Pending', 'Cancelled'], num_rows),
    'CouponUsed': np.random.choice([True, False], num_rows)
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate TotalPrice
df['TotalPrice'] = df['Quantity'] * df['Price']

# Ensure the DataFrame has exactly 10 columns
# If we created more than 10, we'll slice; if less, we'll add dummy columns (not expected here)
if len(df.columns) > num_columns:
    df = df.iloc[:, :num_columns]
elif len(df.columns) < num_columns:
    # This part is unlikely with the current data generation, but included for robustness
    for i in range(len(df.columns), num_columns):
        df[f'ExtraCol{i+1}'] = np.random.randint(0, 100, num_rows)

print(f"Generated DataFrame with {df.shape[0]} rows and {df.shape[1]} columns:")
df
```
